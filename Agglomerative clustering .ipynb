# Agglomerative Hierarchical clustering 

```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#taking a blobs dataset to create clusters 
X,y = make_blobs(n_samples=100,centers=5,random_state=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
data = {'x' : X}
```

```python
df = pd.DataFrame(X)
```

```python
df.head()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
#first we need to find the optimal number of clusters which we can do by forming a dendogram
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(df, method  = "ward"))
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
cluster.fit_predict(df)
```

```python
#storing the fit results and calculating the silhouette_score
cl = cluster.fit_predict(df)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,cl)
```
