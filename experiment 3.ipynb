# Generating samples of gussain distribution 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#
# Draw random samples from normal distribution
# with a predefined mean and standard deviation
#
x = np.random.normal(loc=0, scale = 1, size=1000)
#converting to standard normal distribution
x_mean = np.mean(x);
x_std = np.std(x);
x = (x-x_mean)/x_std
#
# Plot probability distribution function
#
x_sorted = np.sort(x)
plt.figure(figsize=(7, 5))
plt.plot(x_sorted, norm.pdf(x_sorted, loc=0, scale=1))
plt.title("Normal distribution", fontsize=16)
```
